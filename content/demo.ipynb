## Loading data from an API using http_call
- `/proxy/apis` will list all apis available with saved authentication.
- API's with credentials saved in the `/proxy` session are able to be queried easily using `http_call`.
- For prefixes with no saved credentials, the prefix is treated as a https domain to forward the request to.

```python
# List all saved apis
http_call("apis").json()
```

```python
# Credentialed api example
http_call("abuseipdb/api/v2/check", params={"ipAddress": "8.8.8.8"}).json()
```

```python
# Uncredentialed passthrough example
http_call("httpbin.org/get").json()
```
